```python
import spacy

nlp = spacy.load('en')

text = open('punicwars.txt').read()
doc = nlp(text)

for person in set([e.text.strip() for e in doc.ents if e.label_ == 'PERSON']):
    print(person)
```

```python
from spacy.lang.da import Danish

nlp = Danish()

text = """
"Pokémon-Sangen"

Jeg drager ud på livets vej,
Jeg har et enkelt mål,
Pokémon'er fanger jeg,
Min vilje er af stål.

Jeg vil rejse ud i det blå,
Lede overalt,
Og jeg vil træne Pokémon,
Det ' noget jeg har valgt.

Pokémon - du skal fange den!
			Det dig og mig
Vi gør livet til en leg

Pokémon - åhh, vi følges ad,
Og kæmper for verdensfred,

Pokémon - du skal fange den!
			Det si'r sig selv.

Det kræver både mod og held,

Vi må bare hjælpes ad,

Pokémon - du skal fange den!

Jeg skal fange den - Pokémon!
"""

doc = nlp(text)

for entity in doc.ents:
    print(entity.text, entity.label_)
```
