```python
import numpy as np
import pandas as pd
```

```python
# Correlation + Pearson's R
a = np.array([1,1,6,7,2,1,9,8])
b = np.array([110, 120, 850, 870, 99, 101, 907, 745])

# Calculate Pearson product-moment correlation coefficients
corr_coef = numpy.corrcoef(a, b)
# Select Pearson's R
pearsons_r = corr_coef[0,1]
print(pearsons_r)
```
